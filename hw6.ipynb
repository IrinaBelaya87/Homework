```python
import pandas as pd
```

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

data = pd.read_csv('keywords.csv')
```

```python
data.head()
```

```python
def add_region(row):
    spisok = []
    for i in geo_data['Центр']:
        if i in row['keyword']:
            return 'Центр'
    for i in geo_data['Северо-Запад']:
        if i in row['keyword']:
            return 'Северо-Запад'
    for i in geo_data['Дальний Восток']:
        if i in row['keyword']:
            return 'Дальний Восток'
        return 'undefined'
```

```python
data['region'] = data.apply(add_region, axis = 1)

data.groupby('region').count()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
def class_ratings(row):
    if row['rating'] <= 2:
        return 'Низкий рейтинг'
    if row['rating'] <= 4:
        return 'Средний рейтинг' 
    if row['rating'] <= 5:
        return 'Высокий рейтинг'
    else:
        return 'Рейтинг не определен'
```

```python
ratings['class'] = ratings.apply(class_ratings, axis = 1)
ratings.head(10)
```

```python
#Задание 3
```

```python
from datetime import datetime
from datetime import timedelta
import numpy as np
data_list = ratings.groupby(['userId']).count().reset_index()[['userId', 'movieId']]
data_list.head()
```

```python
top_users_100 = data_list[data_list['movieId'] >= 100]
```

```python
top_users_100_list = top_users_100['userId'].tolist()
```

```python
new_data_list = ratings[ ratings['userId'].isin(top_users_100_list) ]
new_data_list.head()
```

```python
new_data_list['timestamp'] = new_data_list['timestamp'].apply(lambda x: datetime.fromtimestamp(x))
```

```python
lifetime_user=new_data_list.groupby('userId')['timestamp'].agg([('Lifetime', lambda x: (np.amax(x)-np.amin(x)  ).days  )])
lifetime_user.sort_values(by='Lifetime', ascending=False).head(10)
```
